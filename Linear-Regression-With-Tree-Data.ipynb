# Linear Regression On Tree Data

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# objective: creating a Linear Regressor for the relationship between girth and volume of trees

df = pd.read_csv("trees.csv")
df
```

```python
df.columns
```

```python
xs = df[" Girth(in)"].tolist()
ys = df[" Volume(ft^3)"].tolist()

plt.scatter(xs, ys)
plt.xlabel('Girth (in)')  
plt.ylabel('Volume (ft^3)')  
plt.title('Data')   
plt.grid(True)
plt.show()
```

```python
# initializing random starting values
theta_0 = 0
theta_1 = 0

# initializing learning rate: trial and error
alpha = 0.0001

# satisfiable error value: trial and error
epsilon = 0.00000001

# training phase
while True:
    derivative_0 = 0
    derivate_1 = 0
    for x in xs:
        derivative_0 += (theta_0 * x + theta_1 - ys[xs.index(x)]) * x
        derivate_1 += (theta_0 * x + theta_1 - ys[xs.index(x)])

    derivative_0 = 2 / len(xs) * derivative_0
    derivate_1 = 2 / len(xs) * derivate_1

    theta_0 -= alpha * derivative_0
    theta_1 -= alpha * derivate_1

    # using Mean Squared Error
    error = 0

    for x in xs:
        if error >= epsilon * .1:
            break
        error += (theta_0 * x + theta_1 - ys[xs.index(x)]) ** 2

    if error <= epsilon:
        break
```

$$MSE=\dfrac{1}{n}\sum_{i=1}^n(Y_i - \hat{Y}_i)^2$$

```python
# the results
print(theta_0, theta_1)
```

```python
func = theta_0 * np.array(xs) + theta_1

plt.scatter(xs, ys)
plt.plot(xs, func, c='r')
plt.xlabel('Girth (in)')  
plt.ylabel('Volume (ft^3)')  
plt.title('Linear Regressor')   
plt.grid(True)
plt.show() 
```
